```python
import pandas as pd
import matplotlib.pyplot as plt
from pyspark.sql import SparkSession

# Initialize Spark session
spark = SparkSession.builder.appName("YouTube Data Analysis").getOrCreate()

# Get the file name from the uploaded file (you will see it in the previous step)
file_name = list(uploaded.keys())[0]

# Load the uploaded file into a PySpark DataFrame
df_spark = spark.read.csv(file_name, header=True, inferSchema=True)

# Convert PySpark DataFrame to Pandas DataFrame
df_pandas = df_spark.toPandas()

# Display the first few rows to inspect the data
df_pandas.head()

# Display column names
df_pandas.columns
```

```python
import matplotlib.pyplot as plt

# Group data by 'category' and sum 'video views'
df_plot = df_pandas[['category', 'video views']].groupby('category').sum().reset_index()

# Create the bar plot
plt.figure(figsize=(12,8))  # Set a larger figure size for better readability
plt.bar(df_plot['category'], df_plot['video views'])
plt.title('YouTube Video Views by Category')
plt.ylabel('Total Video Views')
plt.xlabel('Category')
plt.xticks(rotation=90)  # Rotate x labels for better readability
plt.show()
```

```python
import matplotlib.pyplot as plt

# Group data by 'category' and sum 'video views'
df_plot = df_pandas[['category', 'video views']].groupby('category').sum().reset_index()

# Create the bar plot
plt.figure(figsize=(12,8))  # Set a larger figure size for better readability
plt.bar(df_plot['category'], df_plot['video views'])
plt.title('YouTube Video Views by Category')
plt.ylabel('Total Video Views')
plt.xlabel('Category')
plt.xticks(rotation=90)  # Rotate x labels for better readability
plt.show()
```

```python
# Bar Chart: YouTube Video Views by Category
plt.figure(figsize=(12,8))
plt.bar(df_plot['category'], df_plot['video views'])
plt.title('YouTube Video Views by Category')
plt.ylabel('Total Video Views')
plt.xlabel('Category')
plt.xticks(rotation=90)
plt.show()

```

```python
# Bar Chart: YouTube Video Views by Category
plt.figure(figsize=(12,8))
plt.bar(df_plot['category'], df_plot['video views'])
plt.title('YouTube Video Views by Category')
plt.ylabel('Total Video Views')
plt.xlabel('Category')
plt.xticks(rotation=90)
plt.show()
```

```python
# Pie Chart: Distribution of YouTubers by Category
df_category_count = df_pandas['category'].value_counts()

plt.figure(figsize=(10,8))
plt.pie(df_category_count, labels=df_category_count.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of YouTubers by Category')
plt.axis('equal')  # Equal aspect ratio ensures the pie is drawn as a circle.
plt.show()
```

```python
# Histogram: Distribution of Subscribers
plt.figure(figsize=(10,8))
plt.hist(df_pandas['subscribers'], bins=20, color='blue', alpha=0.7)
plt.title('Distribution of Subscribers')
plt.xlabel('Number of Subscribers')
plt.ylabel('Frequency')
plt.show()
```

```python
# Scatter Plot: Subscribers vs. Video Views
plt.figure(figsize=(10,8))
plt.scatter(df_pandas['subscribers'], df_pandas['video views'], alpha=0.5)
plt.title('Subscribers vs. Video Views')
plt.xlabel('Subscribers')
plt.ylabel('Video Views')
plt.show()
```

```python
# Line Plot: Lowest and Highest Monthly Earnings by Category
df_earnings = df_pandas[['category', 'lowest_monthly_earnings', 'highest_monthly_earnings']].groupby('category').sum().reset_index()

plt.figure(figsize=(12,8))
plt.plot(df_earnings['category'], df_earnings['lowest_monthly_earnings'], label='Lowest Monthly Earnings', marker='o')
plt.plot(df_earnings['category'], df_earnings['highest_monthly_earnings'], label='Highest Monthly Earnings', marker='o')
plt.title('Monthly Earnings by Category')
plt.ylabel('Earnings (in USD)')
plt.xlabel('Category')
plt.xticks(rotation=90)
plt.legend()
plt.show()
```
